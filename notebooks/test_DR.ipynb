```python
# -- load in file and get info
loaded_file = np.load(one_file, allow_pickle=True)
loaded_arrays = {
        key: loaded_file[key] for key in loaded_file.files
    }
print(loaded_arrays)
```

```python
# -- these are the points or waveform features identified as blips in the event
features_set = loaded_arrays['features']
print(len(features_set[0]))
print(features_set[0].shape)
```

```python
# -- the cluster_id is the cluster label given by DBScan (to each point in the original features array (chan,tdc,adc))
# -- cluster_id's are sequential
# -- this represents how DBSCAN "clusters" the points(or features) into clusters
cluster_set = loaded_arrays['10.0_6_cluster_ids']
print(cluster_set.shape)
print(cluster_set)
print(len(cluster_set[0]))
```

```python
# -- cluster_indices are the indices of the data_{ii}.pt files associated w/ each cluster_id from each event
# -- this represents the mapping from cluster_id to pytorch (pre-processed) files.
# -- Note: these are numbered uniquely across the entire training dataset,
#    so the pt_index picks up where the previous event left off.
cluster_ind_set = loaded_arrays['10.0_6_cluster_indices']
print(f"The length of the cluster_ind_set for evt=1 is : {len(cluster_ind_set[0])}")
print(f"The shape for the cluster_ind_set for evt=1 is : {cluster_ind_set.shape}")
print(cluster_ind_set)
```

```python
# -- these are the prediction (probabilities) for each particle type
# -- There is one array for each cluster (data_ii.pt file) w/ one entry per particle class defined
particle_set=loaded_arrays['10.0_6_particle']
print(len(particle_set[0]))
print(particle_set[0].shape)
print(particle_set)
```

```python
# -- add up the probabilities to make sure
i=0
totals=[]
for probs in particle_set[0]:
    total = np.sum(probs)
    totals.append(total)
    if (i==0):
        print(f"total for first array = {total}")
    
    i+=1
    
```
